```python
#basic function without parameter

def greet():
  print("hello,welcome to blearn")

greet()
```

```python
#basic funciton with parameters

def greet(name):
  print("hello",name,"welcome to blearn")

greet("adarsh")  #positional arguments
greet(name = "amar")  #keyword arguments
```

```python
#function with two parameters

def respond(name,age):
  print("hello",name,"and he is",age,"years old")

respond("abhi",26) #posititonal parameter
respond(26,"abhi")
```

```python
#keyword parameter

def respond(name,age):
  print("hello",name,"and he is",age,"years old")

respond(name="abhi",age=22)
```

```python
#default parameter

def respond(name="amar",age=26):
  print("hello",name,"and he is",age,"years old")

respond("abhi",30)

respond()
```

```python
#sunm of list

def sum_of_list(num):
  return sum(num)
mylist = [1,2,3]
result = sum_of_list(mylist)
print(result)
```

```python
#perimeter of a rectangle

def perime(a,b):
  return 2*(a+b)
a=2
b=3
result = perime(a,b) #perime(2,3)
print(result)
```

```python
#exponenitial

def power(base,exponent):
  if exponent == 0:
    return 1
  return base * power(base, exponent-1)
print(power(2,3))
```

```python
#minimum

def findmini(num):
  return min(num)
print(findmini([4,2,1,5]))
```

```python
#maximum

def maximum(num):
  return max(num)
m=[2,5,6,20]
result=maximum(m)
print(result)
```

```python
#kilometers to meters

def km_to_m(num):
  return num * 1000
print(km_to_m(5),"m")
```

```python
#fahrenheit to celsius

def fahren(f):
  c=(f-32)*5/9
  return c
print(fahren(102))
```

```python
#count vowels

def count_vowels(text):
  vowel="aeiouAEIOU"
  count=0
  for char in text:
    if char in vowel:
      count+=1
  return count

user_input=input("enter your name: ")
print("number of vowels")
count_vowels(user_input)
```

```python
# Map syntax

# map(fucntion, listname)
```

```python
#map()

num=[1,2,3,4]
squares = list(map(lambda x:x*x, num))
print(squares)
```

```python
#celsius to fahrenheit using lambda fucniton

celsius = [0.0, 10.0, 20.0, 30.0]
fahren  = list( map( lambda x: (x*9/5)+32,celsius )) #map(function,list name)
print(fahren)
```
